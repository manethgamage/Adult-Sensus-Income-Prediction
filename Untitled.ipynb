```python
Q1 = data['fnlwgt'].quantile(0.25)
Q3 = data['fnlwgt'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outlier_data = data[(data['fnlwgt'] < lower_bound) | (data['fnlwgt'] > upper_bound)]

print("Shape of outlier data:", outlier_data.shape)

median_fnlwgt = data['fnlwgt'].median()

data.loc[(data['fnlwgt'] < lower_bound) | (data['fnlwgt'] > upper_bound), 'fnlwgt'] = median_fnlwgt
```

```python
data['marital_status'] = data['marital_status'].replace([' Divorced',' Married-spouse-absent',' Never-married',' Separated',' Widowed'],'Single')
data['marital_status'] = data['marital_status'].replace([' Married-AF-spouse',' Married-civ-spouse'],'Couple')
data.head(10)
```

```python
def f(x):
    if x['workclass'] == ' Federal-gov' or x['workclass']== ' Local-gov' or x['workclass']==' State-gov': return 'govt'
    elif x['workclass'] == ' Private':return 'private'
    elif x['workclass'] == ' Self-emp-inc' or x['workclass'] == ' Self-emp-not-inc': return 'self_employed'
    else: return 'without_pay'
    
    
data['workclass']=data.apply(f, axis=1)

data.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['workclass'] = le.fit_transform(data['workclass'])
data['marital_status'] = le.fit_transform(data['marital_status'])
data['occupation'] = le.fit_transform(data['occupation'])
data['relationship'] = le.fit_transform(data['relationship'])
data['race'] = le.fit_transform(data['race'])
data['sex'] = le.fit_transform(data['sex'])
data['native_country'] = le.fit_transform(data['native_country'])
data['income'] = le.fit_transform(data['income'])
# data['capital_gain_category'] = le.fit_transform(data['capital_gain_category'])
data.head()
```

```python
from sklearn.naive_bayes import GaussianNB
gb = GaussianNB()
rf = RandomForestClassifier(bootstrap=False, max_depth=30, max_features='log2',
                       min_samples_leaf=8, min_samples_split=20,
                       n_estimators=450, random_state=2)
gb.fit(X_train,Y_train)
rf.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score 
y_pred = gb.predict(X_test)
y_preds = rf.predict(X_test)
y_predd_nb = gb.predict(X_train)
y_predd_rd = rf.predict(X_train)

print(accuracy_score(Y_test,y_pred)*100)
print("train:naive ",accuracy_score(Y_train,y_predd_nb)*100)
print("train rf ",accuracy_score(Y_train,y_predd_rd)*100)
print(accuracy_score(Y_test,y_preds)*100)
```

```python
from sklearn.metrics import precision_score, recall_score

precision = precision_score(Y_test, y_pred)

# Calculate recall
recall = recall_score(Y_test, y_pred)

# Print precision and recall
print("Precision:", precision)
print("Recall:", recall)
```

```python
from sklearn.metrics import f1_score, confusion_matrix

# Assuming you have true labels (y_true) and predicted labels (y_pred)
# Compute the F1 score
f1 = f1_score(Y_test, y_pred)

# Compute the confusion matrix
cm = confusion_matrix(Y_test, y_pred)

# Print the F1 score
print("F1 Score:", f1)

# Print the confusion matrix
print("Confusion Matrix:")
print(cm)
```
